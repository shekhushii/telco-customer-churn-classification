```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv(r"C:\Users\khush\Downloads\archive\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df.dropna(how='any' , inplace= True)
```

```python
X = df.drop(['customerID','Churn'] , axis=1) 
y=df.Churn.values
```

```python
X
```

```python
y
```

```python
# convert categorcial attributes to numerical attribute -- > feature encoding
X = pd.get_dummies(X,columns = ['gender', 'Partner', 'Dependents','PhoneService',
        'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'],drop_first = True)
```

```python
bool_cols = X.select_dtypes(include='bool').columns
X[bool_cols] = X[bool_cols].astype(int)
```

```python
X.head(1)
```

```python
# splitting the data into traing nd testing 
from sklearn.model_selection import train_test_split
X_train,X_test ,y_train,y_test  = train_test_split(X,y,test_size=0.25)
```

```python
# feature scaling :
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc
```

```python
X_test_sc
```

```python
## KNN  CLASSIFIER  
from sklearn.neighbors import KNeighborsClassifier

# initiating my classifier 
model = KNeighborsClassifier()

# passing the value to classifier
model.fit(X_train_sc,y_train)
```

```python
 y_pred = model.predict(X_test_sc)
```

```python
# classification metririx 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
## decison tree classifier  
from sklearn.tree import DecisionTreeClassifier 
#initiating the classifier 
model_dt = DecisionTreeClassifier()  
#passing the data to classifier 
model_dt.fit(X_train_sc,y_train)
```

```python
y_pred_dt = model_dt.predict(X_test_sc)
```

```python
# classification metririx 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred_dt)*100)
```

```python
# random forest classifier 
from sklearn.ensemble import RandomForestClassifier

# initiating the classifier
model = RandomForestClassifier(n_estimators=200)

# passing the data to classifier
model.fit(X_train_sc,y_train)
```

```python
y_pred_rn = model.predict(X_test_sc)
```

```python
# classification metririx 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred_rn)*100)
```

```python
from sklearn.metrics import classification_report 
knn_report  = classification_report(y_test,y_pred)
dt_report = classification_report(y_test,y_pred_dt)
rn_report = classification_report(y_test,y_pred_rn)
```

```python
 # NAIVE BAYES CLASSIFIER
from sklearn.naive_bayes import BernoulliNB 

# initiating my classifier
model_nb = BernoulliNB()

# passing the data to classifier
model_nb.fit(X_train_sc,y_train) 

# prediction
y_pred_nb = model_nb.predict(X_test_sc)

# accuracy 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred_nb)*100)
```

```python
nb_report = classification_report(y_test,y_pred_nb)
print(nb_report)
```

```python
# SVM CLASSIFIER 
from sklearn.svm import SVC 

# initiating my classifier
model_svc = SVC()

# passing the data to classifier
model_svc.fit(X_train_sc,y_train) 

# prediction
y_pred_svc = model_svc.predict(X_test_sc)

# accuracy 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred_svc)*100)
```

```python
svc_report = classification_report(y_test,y_pred_svc)
print(svc_report)
```

```python
# LOGISTIC REGRESSION 
from sklearn.linear_model import LogisticRegression

# initiating my classifier
model_lr = LogisticRegression()

# passing the data to classifier
model_lr.fit(X_train_sc,y_train) 

# prediction
y_pred_lr = model_lr.predict(X_test_sc)

# accuracy 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred_lr)*100)
```

```python
lr_report = classification_report(y_test,y_pred_lr)
print(lr_report)
```
